```python
df = pd.merge(df_summary[['tripId','device','startlocaltime',
                         'endlocaltime','startlatitude','startlongitude',
                         'endlatitude','endlongitude','numBSMTx', 'numGNSSOutages', 'numCANOutages',
                         'numNormalBSMRx', 'numShadowBSMRx',
                         'numSPATRx', 'numWarnings', 'numInforms', ]], 
              df_host[['tripId','speed','speedMph','filedate','filetimeutc',
                      'localTimeMs','latitude','longitude','gpsspeed','elevation',
                      'brakestatus','turnsignal','yawrate','longaccel']], on = 'tripId')
```

```python
df.shape
```

```python
df.head()
```

```python
df = df.sort_values(by = ['tripId','filedate','localTimeMs'])
df.head()
```

```python
df['speedMph'] = df['speedMph'].apply(lambda x: x if x < 100 else np.nan)
```

```python
df.head()
```

```python
df['speedMph'].isna().mean()*100
```

```python
df['speedMph'].describe()
```

```python
df['yawrate'] = df['yawrate'].apply(lambda x: x if abs(x) < 327.67 else np.nan)
```

```python
df['yawrate_speed'] = abs(df['yawrate'])*df['speedMph']
```

```python
df['yawrate_speed'].describe()
```

```python
df.sort_values(by = 'yawrate_speed', ascending = False).head()
```

```python
df.head()
```

```python
df['speedDiff'] = df.groupby('tripId')['speedMph'].diff()
df['timediff'] = df.groupby('tripId')['localTimeMs'].diff()
df['timediff'] = df['timediff']/1000
```

```python
df['speedDiff_ms'] = df['speedDiff']*0.44704
```

```python
df['speedrate_ms2'] = df['speedDiff_ms']/(df['timediff'])
```

```python
df['speedrate_ms2'].describe()
```

```python
df.sort_values(by = 'speedrate_ms2',ascending = False).head()
```

```python
df.loc[df['timediff'] < 0.1, 'speedrate_ms2'] = np.nan
```

```python
df['speedrate_ms2'].describe()
```

```python
df[df['speedrate_ms2'] > 1.5].shape
```

```python
df[df['speedrate_ms2'] < -1.5].shape
```

```python
df['sudden_acceleration'] = df['speedrate_ms2'] > 1.5
df['sudden_decceleration'] = df['speedrate_ms2'] < -1.5
```

```python
df['yawrate'].describe()
```

```python
df['accelerated_turn'] = (df['sudden_acceleration'] & (abs(df['yawrate']) > 30)) 
```
