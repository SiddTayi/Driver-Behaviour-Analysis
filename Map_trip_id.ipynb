```python
# we dont have a unique value for each trip
# we will extract the device id, start time and end time from summary table and
# map the trips to a unique trip id

import uuid


count = 0
summary_table_trail['trip_uuid'] = 0
host_table_trail['trip_uuid'] = 0
for i in summary_table_trail.index:
  random_uuid = uuid.uuid4()

  print(summary_table_trail['device'][i], summary_table_trail['startlocaltime'][i], summary_table_trail['endlocaltime'][i])

  device_condition = host_table_trail['device'] == summary_table_trail['device'][i]
  time_condition = host_table_trail['localtimems'].between(summary_table_trail['startlocaltime'][i], summary_table_trail['endlocaltime'][i])


  host_table_trail.loc[device_condition & time_condition, 'trip_uuid'] = random_uuid
  summary_table_trail['trip_uuid'][i] = random_uuid
  result = host_table_trail[device_condition & time_condition]
  print(result.shape)

  count+=1
  if(count == 2):
    break
```
